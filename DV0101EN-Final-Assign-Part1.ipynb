####  TASK 1.2: Plot different lines for categories of vehicle type and analyse the trend to answer the question Is there a noticeable difference in sales trends between different vehicle types during recession periods?

````python
import matplotlib.pyplot as plt

# Group by Year and Vehicle_Type and sum sales
vehicle_sales_by_year = df.groupby(['Year', 'Vehicle_Type'])['Automobile_Sales'].sum().unstack()

# Plotting
plt.figure(figsize=(12, 6))
vehicle_sales_by_year.plot(marker='o', figsize=(12, 6))

plt.title('Automobile Sales by Vehicle Type Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(vehicle_sales_by_year.index, rotation=45)
plt.grid(True)
plt.legend(title='Vehicle Type')
plt.tight_layout()

# Highlight recession years with vertical shading
recession_years = df[df['Recession'] == 1]['Year'].unique()
for year in recession_years:
    plt.axvline(x=year, color='gray', linestyle='--', linewidth=0.8, alpha=0.5)

plt.show()
````

### TASK 1.3: Use the functionality of **Seaborn Library** to create a visualization to compare the sales trend per vehicle type for a recession period with a non-recession period.


<details><summary>Click here for a hint</summary>

<p>
     To visualize the average number of vehicles sold during recession and non-recession periods, you can use a bar chart

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter recession and non-recession data
recession_df = df[df['Recession'] == 1]
non_recession_df = df[df['Recession'] == 0]

# Set up the plot size
plt.figure(figsize=(14, 6))

# Create lineplot for recession period
sns.lineplot(data=recession_df, x='Year', y='Automobile_Sales', hue='Vehicle_Type', marker='o')
plt.title('Automobile Sales by Vehicle Type During Recession')
plt.xlabel('Year')
plt.ylabel('Automobile Sales')
plt.legend(title='Vehicle Type')
plt.grid(True)
plt.show()

# Create lineplot for non-recession period
plt.figure(figsize=(14, 6))
sns.lineplot(data=non_recession_df, x='Year', y='Automobile_Sales', hue='Vehicle_Type', marker='o')
plt.title('Automobile Sales by Vehicle Type During Non-Recession')
plt.xlabel('Year')
plt.ylabel('Automobile Sales')
plt.legend(title='Vehicle Type')
plt.grid(True)
plt.show()
````

### TASK 1.4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, you want to find more insights from the data to understand the reason. <br>Plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods? 
<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


````python
import matplotlib.pyplot as plt

# Create a figure and add subplots
fig = plt.figure(figsize=(14, 6))

# First subplot: GDP during recession periods
ax1 = fig.add_subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
recession_data = df[df['Recession'] == 1]  # Filter data for recession period
ax1.plot(recession_data['Date'], recession_data['GDP'], color='red', label='GDP during Recession')
ax1.set_title('GDP during Recession Period')
ax1.set_xlabel('Date')
ax1.set_ylabel('GDP')
ax1.tick_params(axis='x', rotation=45)
ax1.grid(True)
ax1.legend()

# Second subplot: GDP during non-recession periods
ax2 = fig.add_subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
non_recession_data = df[df['Recession'] == 0]  # Filter data for non-recession period
ax2.plot(non_recession_data['Date'], non_recession_data['GDP'], color='green', label='GDP during Non-Recession')
ax2.set_title('GDP during Non-Recession Period')
ax2.set_xlabel('Date')
ax2.set_ylabel('GDP')
ax2.tick_params(axis='x', rotation=45)
ax2.grid(True)
ax2.legend()

# Adjust layout to make sure there's no overlap
plt.tight_layout()

# Show the plot
plt.show()
````

### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python
import matplotlib.pyplot as plt

# Filter for non-recession years
non_recession_data = df[df['Recession'] == 0]

# Create a Bubble Plot
plt.figure(figsize=(10, 6))

# Plot the bubble chart
plt.scatter(non_recession_data['Month'], 
            non_recession_data['Automobile_Sales'], 
            s=non_recession_data['Seasonality_Weight']*1000,  # Bubble size, scale the seasonality weight
            c=non_recession_data['Seasonality_Weight'],  # Bubble color based on seasonality weight
            cmap='coolwarm',  # Color map
            alpha=0.6,  # Transparency of the bubbles
            edgecolors='w',  # White edges for better contrast
            label='Seasonality Impact')

# Customize plot
plt.title('Seasonality Impact on Automobile Sales')
plt.xlabel('Month')
plt.ylabel('Automobile Sales')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.colorbar(label='Seasonality Weight')  # Color bar to show the scale of seasonality weight
plt.grid(True)

# Show the plot
plt.show()
````

 ### TASK 1.9: Develop a lineplot to analyse the effect of the unemployment rate on vehicle type and sales during the Recession Period.
 <br>Analyze the effect of the unemployment rate on vehicle type and sales during the Recession Period
 #### You can create a lineplot and title the plot as 'Effect of Unemployment Rate on Vehicle Type and Sales'


<details><summary>Click here for a hint</summary>

<p>
    Filter out the data for recession period<br>
     Make use of lineplot() from seaborn and pass the relavent data</p>
</details>


````python
import seaborn as sns

# Filter data for recession period (Recession == 1)
df_recession = df[df['Recession'] == 1]

# Create a line plot to analyze the effect of unemployment rate on vehicle type and sales
plt.figure(figsize=(12, 8))
sns.lineplot(data=df_recession, 
             x='Date', 
             y='unemployment_rate', 
             hue='Vehicle_Type', 
             marker='o', 
             linewidth=2)

# Adding a secondary axis to show Automobile Sales for comparison
ax2 = plt.gca().twinx()  # Create a secondary y-axis
sns.lineplot(data=df_recession, 
             x='Date', 
             y='Automobile_Sales', 
             ax=ax2, 
             color='gray', 
             linestyle='--', 
             label='Automobile Sales', 
             linewidth=2)

# Set plot title and labels
plt.title('Effect of Unemployment Rate on Vehicle Type and Sales during Recession Period', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Unemployment Rate (%)', fontsize=12)
ax2.set_ylabel('Automobile Sales', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the legend
plt.legend(title='Vehicle Type', loc='upper left', fontsize=10)

# Show the plot
plt.show()
````
